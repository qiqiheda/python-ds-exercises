***Ex3.*** Using a single line of code, retrieve the average age of female and male possum (hint: use group_by)

```python
#DataFrame.groupby(by=None, axis=0, level=None, as_index=True, sort=True, group_keys=True, observed=False, dropna=True)[source]
df.groupby('sex')['age'].mean()
```

***Ex4.*** Output the percentage of male and female. hint: use value_counts()

```python
# Series.value_counts(normalize=False, sort=True, ascending=False, bins=None, dropna=True)[source]
# Return a Series containing counts of unique values.

# df['sex'].value_counts() # m 61 f 43

# setting the normalize parameter to True to get proportions instead of counts. 
# mul(100): multiply by 100 to get percentage format
# round decimal to 2 digit
# astype(str): convert data type to string to prepare for concat
# add % symbol
df['sex'].value_counts(normalize=True).mul(100).round(2).astype(str) + '%'
```

***Ex12.*** Sort by decreasing age and increasing tail and keep records 5 (included) to 50 (not included)

```python
# DataFrame.sort_values(by, *, axis=0, ascending=True, inplace=False, kind='quicksort', 
#                       na_position='last', ignore_index=False, key=None)
# we can sort multiple columns, wrap the labels in [].
# The ascending parameter is set to [False, True] to specify the sorting order for each column.
sort_result = df.sort_values(by=['age', 'taill'], ascending=[False, True]).iloc[5:50, :]
sort_result.head()
```
